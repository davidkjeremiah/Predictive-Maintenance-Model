# Data Exploratipn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the raw data
df = pd.read_csv('../data/raw/machine failure.csv')

# Explore the data structure
print(df.info())
print(df.describe())
```

```python
# Visualize the feature distributions
plt.figure(figsize=(12, 8))
sns.histplot(df['Air temperature [K]'])
plt.title('Distribution of Air Temperature')
plt.show()
```

```python
# Analyze the relationship between features and the target
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Rotational speed [rpm]', y='Torque [Nm]', hue='Machine failure', data=df)
plt.title('Rotational Speed vs Torque')
plt.show()
```
